### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style='ticks')
```

### Step 2. Create the DataFrame that should look like the one below.

```python
raw = {'first_name' : ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
       'last_name' : ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
       'age' : [42, 52, 36, 24, 73],
       'female' : [0, 1, 1, 0, 1],
       'preTestScore' : [4, 24, 31, 2, 3],
       'postTestScore' : [25, 94, 57, 62, 70]}
df = pd.DataFrame(data=raw)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(x=df['preTestScore'], y=df['postTestScore'], s=df['age'])
plt.title('post- and preTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
def female_color(female):
    if female:
        return 'pink'
    return 'blue'

plt.scatter(x=df['preTestScore'], y=df['postTestScore'], s=df['postTestScore'] * 4.5, c=df['female'].apply(female_color))
plt.title('post- and preTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

### BONUS: Create your own question and answer it.

```python
df['avgScore'] = (df['postTestScore'] + df['preTestScore']) / 2
plt.scatter(x=df['age'], y=df['postTestScore'], s=df['avgScore'], c=df['female'].apply(female_color))
plt.title('avgScore by age')
plt.xlabel('Age')
plt.ylabel('avgScore')
plt.show()
```
